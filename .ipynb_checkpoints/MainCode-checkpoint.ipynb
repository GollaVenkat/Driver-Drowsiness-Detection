```python
import cv2
import os
from keras.models import load_model
# from model import predict_classes
import numpy as np
from pygame import mixer
import time
```

```python
mixer.init()
# this sound will play when drowsiness is deteched
sound = mixer.Sound('alarm.wav')
# These classfifiers are used to detect ROI (Face) ,left eye and right eye
face = cv2.CascadeClassifier('haar cascade files\haarcascade_frontalface_alt.xml')
leye = cv2.CascadeClassifier('haar cascade files\haarcascade_lefteye_2splits.xml')
reye = cv2.CascadeClassifier('haar cascade files\haarcascade_righteye_2splits.xml')



lbl=['Close','Open']
# This is our CNN model
model = load_model('models/cnncat2.h5')
path = os.getcwd()
# It is usally represents camera which reads frames from camera
cap = cv2.VideoCapture(0)
font = cv2.FONT_HERSHEY_COMPLEX_SMALL
# on based of score we will detect whether there is drowsiness or not
score=0
thicc=2
rpred=[99]
lpred=[99]
# We will have to take every frame from camera so for that we need a loop
while(True):
#     cap.read() return tupe containing 2 value ret will tell us whether its succesfully
# able to read frame or not and other value is frame
    ret, frame = cap.read()
    height,width = frame.shape[:2] 
#     Converting color image to gray scale image
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
#     detecting face left and right eye from gray scaled image
    faces = face.detectMultiScale(gray,minNeighbors=5,scaleFactor=1.1,minSize=(25,25))
    left_eye = leye.detectMultiScale(gray)
    right_eye =  reye.detectMultiScale(gray)
    cv2.rectangle(frame, (0,height-50) , (200,height) , (0,0,0) , thickness=cv2.FILLED )
# drawing a rectangle box on face
    for (x,y,w,h) in faces:
        cv2.rectangle(frame, (x,y) , (x+w,y+h) , (100,100,100) , 1 )
# checking whether right is closed or not 
    for (x,y,w,h) in right_eye:
#         from the frame i just selected right eye pixels
        r_eye=frame[y:y+h,x:x+w]
#     converting extracted right eye to gray scale image
        r_eye = cv2.cvtColor(r_eye,cv2.COLOR_BGR2GRAY)
#     i did some preprocessing of image like converted to 24x24 shape and normalized every pixel 
# so it will be in 0-1
        r_eye = cv2.resize(r_eye,(24,24))
        r_eye= r_eye/255
        r_eye=  r_eye.reshape(24,24,-1)
        r_eye = np.expand_dims(r_eye,axis=0)
#         Predicted wheyther right is closed or not using cnn model
        rpred = np.argmax(model.predict(r_eye),axis=-1)
        if(rpred[0]==1):
            lbl='Open' 
        if(rpred[0]==0):
            lbl='Closed'
        break

    for (x,y,w,h) in left_eye:
#         from the frame i just selected left eye pixels
        l_eye=frame[y:y+h,x:x+w]
#     converting extracted right eye to gray scale image
        l_eye = cv2.cvtColor(l_eye,cv2.COLOR_BGR2GRAY)  
#     i did some preprocessing of image like converted to 24x24 shape and normalized every pixel 
# so it will be in 0-1
        l_eye = cv2.resize(l_eye,(24,24))
        l_eye= l_eye/255
        l_eye=l_eye.reshape(24,24,-1)
        l_eye = np.expand_dims(l_eye,axis=0)
#         Predicted wheyther right is closed or not using cnn model
        lpred = np.argmax(model.predict(l_eye),axis=-1)
        if(lpred[0]==1):
            lbl='Open'   
        if(lpred[0]==0):
            lbl='Closed'
        break
#     if both eyes closed then we increasing score else decreasing it
    if(rpred[0]==0 and lpred[0]==0):
        score=score+1
        cv2.putText(frame,"Closed",(10,height-20), font, 1,(255,255,255),1,cv2.LINE_AA)
    else:
        score=score-1
        cv2.putText(frame,"Open",(10,height-20), font, 1,(255,255,255),1,cv2.LINE_AA)
    
#    Score wont be negative     
    if(score<0):
        score=0   
#         writing score, whether eye is opened or closed on images
    cv2.putText(frame,'Score:'+str(score),(100,height-20), font, 1,(255,255,255),1,cv2.LINE_AA)
    if(score>15):
        #person is feeling sleepy so we beep the alarm
        cv2.imwrite(os.path.join(path,'image.jpg'),frame)
        try:
            sound.play()
            
        except:  # isplaying = False
            pass
        if(thicc<16):
            thicc= thicc+2
        else:
            thicc=thicc-2
            if(thicc<2):
                thicc=2
        cv2.rectangle(frame,(0,0),(width,height),(0,0,255),thicc) 
    cv2.imshow('frame',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
